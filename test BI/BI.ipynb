```python
import os
import pandas as pd
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```python
path = os.path.dirname(os.path.abspath(__name__)) + '\\Crypto BOT\\save'
dta = pd.read_csv(path + '\\series.csv')
dta = dta.drop(dta.columns[0], axis = 1).set_index('date')
not_save = dta.columns.copy()
display(dta.head(3))
```

```python
from scipy.stats import gaussian_kde

yields = ((dta['price']/dta['price'].iloc[0])-1)

kde = gaussian_kde(yields.values, bw_method = 1)  # bw_method specifies the bandwidth
x = np.linspace(-0.05, 0.05, 1000)
y = kde(x)

plt.plot(x, y)
plt.title("Kernel Density Estimate")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
